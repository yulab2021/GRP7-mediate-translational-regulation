```r
# Supplementary Fig. 5g

library(data.table)
library(stringr)
library(clusterProfiler)
library(org.At.tair.db)
library(ggplot2)

Col_Mock_ABA_TE_only_gene_down = setdiff(rownames(Col0_ABAvsMock_sig_down), rownames(g7g8_ABAvsMock_sig_down))
length(Col_Mock_ABA_TE_only_gene_down)

Col_Mock_ABA_TE_only_gene_up = setdiff(rownames(Col0_ABAvsMock_sig_up), rownames(g7g8_ABAvsMock_sig_up))
length(Col_Mock_ABA_TE_only_gene_up)

t2n <- fread("data/t2n-zj.txt",header = F,sep = "\t")
t2g <- fread("data/t2g-zj.txt",header = F,sep = "\t")

mydf <- data.frame(Entrez = c(Col_Mock_ABA_TE_only_gene_up, Col_Mock_ABA_TE_only_gene_down),
                   group = c(rep("up", length(Col_Mock_ABA_TE_only_gene_up)),rep("down", length(Col_Mock_ABA_TE_only_gene_down))))


compare_result <- compareCluster(Entrez ~ group, data = mydf,fun = "enricher",TERM2GENE = t2g,TERM2NAME = t2n,pvalueCutoff = 0.05)
compare_result@compareClusterResult$Cluster <- factor(compare_result@compareClusterResult$Cluster, levels = c("up", "down"))
#write.table(compare_result@compareClusterResult, "Col_Mock_ABA_TE_only_gene_up_down_GO.txt", sep = "\t", quote = F, row.names = F)

compareClusterResult <- read.table("data/Col_Mock_ABA_TE_only_gene_up_down_GO_1.txt",header = T,sep = "\t")
compare_result@compareClusterResult <- compareClusterResult
compare_result@compareClusterResult$Cluster <- factor(compare_result@compareClusterResult$Cluster, levels = c("up", "down"))

#pdf("fig5_Col_Mock_ABA_TE_only_gene_up_down_GO_1.pdf",width = 5,height = 7)
dotplot(compare_result, showCategory = 20) +
  theme_minimal() +
  theme(panel.border = element_rect(color = "black", fill = NA, linewidth = 0.3), axis.text.y = element_text(size = 12), axis.text.x = element_text(size = 12)) +
  xlab(NULL)+
  scale_y_discrete(labels=function(y) str_wrap(y, width=100))
#dev.off()
```
